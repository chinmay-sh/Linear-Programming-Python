# Creating Graph of Lines and Feasible Region

Graph Setup

```
# axis range setup
range=2000

# figure setup
axis = np.arange(range)
fig = go.Figure(layout={'xaxis':{'range': [-50, range]},'yaxis':{'range':[-50,range]}})
fig.add_trace(go.Scatter(x=axis, y=(axis*0),name='x-axis',line={'color':'#000'}))
fig.add_trace(go.Scatter(x=axis*0, y=axis,name='y-axis',line={'color':'#000'}))

# Adding equation lines
count = 0
for i in eqs:
  count +=1
  x = np.arange(i[2]/i[0])
  fig.add_trace(go.Scatter(x=x, y=((i[2]-i[0]*x)/i[1]),name='Eq '+str(count),hovertext=(str(i[0])+'x + '+str(i[1])+'y = '+str(i[2]))))
fig.show()
```
